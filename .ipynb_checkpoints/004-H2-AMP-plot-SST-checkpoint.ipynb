## Introduction to xarray,
## which "makes working with labelled multi-dimensional arrays in Python simple, efficient, and fun!"

I have downloaded a day of sea surface temperature data into a 'data' folder 

```python
ncfile='data/20220228120000-C3S-L4_GHRSST-SSTdepth-OSTIA-GLOB_ICDR2.1-v02.0-fv01.0.nc'
```

```python
import xarray as xr
ds=xr.open_dataset(ncfile)
```

If the above does not work you may need to install a nectDF dependency for xarray.

The coordinates of a netcdf file are the labels of the data variables are organised in time and space. We can eithr click on the ds table above to explore or we can type `ds.coords` to see the coordinates of our data. Dimensions, on the other hand are the realy physiocal dimensions of the dataset and have a name and length only

```python
ds.coords
```

This data set has the coordinates of time, latitude, and longitude. 

We can ploty the data using the `plot` option.

```python
ds.analysed_sst.plot()
```

```python
da=ds.analysed_sst.sel(lat=slice("40", "60"),lon=slice("-20", "0"))
da.plot()
```

```python
ds.analysed_sst.loc["2022-02-28","40":"60", "-180":"0"].plot()
```
